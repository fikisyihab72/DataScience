<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 

Perbedaan Series dengan list dan dictionary
- List jika di print hanya menghasilkan valuenya saja tapi kalau Series jika di print akan menampilkan index dan valuenya
- Dalam Series index tidak harus bertipe integer. dapat diganti dengan tipe data lain
- Dictionary tidak bisa dilakukan Slicing, sedangkan Series bisa  
- Series bisa mempuyai Key duplikat

Perbedaan DataFrame dan Series
- Series adalah struktur data yang hanya memiliki 1 dimensi (1D). berisi index dan value
- DataFrame adalah struktur data yang memiliki 2 atau lebih dimensi (multi-dimensi). berisi index row, column, dan value


```python
# code here
obj_series = pd.Series(['Hanzo', 'Dva', 'Doomfist', 'Zenyatta'], 
                       index=['a','b','c','d'])

print(obj_series)
```

Jawab: 
Statistika deskriptif merupakan suatu metode yang berhubungan dengan pengumpulan atau penyajian data sampai memberi informasi yang berguna.

Statistika di golongkan menjadi dua bagian, yakni statistika deskriptif dan statistika inferensia yang dilakukan berdasarkan aktivitas yang dilakukan.

Untuk contoh dari statistika deskriptif sendiri antara lain yaitu:

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here

print('nilai rata-rata dari column price       :', diamond['price'].mean())
print('nilai standar deviasi dari column depth : ', diamond['depth'].std())
print('Berapa nilai maximum dari column carat  :', diamond['carat'].max())

display(diamond.describe())
```
